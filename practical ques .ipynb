```python
# 1.
import thinkplot as d
import thinkdsp as t
from IPython.display import Audio
x = t.SinSignal(freq = 10, amp = 5, offset = 0)
x_wave = x.make_wave(duration = 1, start = 0, framerate = 42000)
x_s1 = x_wave.make_spectrum()

y = t.SinSignal(freq = 1, amp = 5, offset = 0)
y_wave = y.make_wave(duration = 1, start = 1.5, framerate = 42000)
y_s2 = y_wave.make_spectrum()

z = t.SinSignal(freq = 2000, amp = 5, offset = 0)
z_wave = z.make_wave(duration = 1,start = 3,framerate = 42000)
z_s3 = z_wave.make_spectrum()

x_s1.plot()
```

```python
# 5
import thinkdsp as t
ab=t.SinSignal(freq=330,amp=1,offset=0)
ab_wave=ab.make_wave(duration=1.5,start=1.5,framerate=48000)
b=t.SinSignal(freq=220,amp=1,offset=0)
b_wave=b.make_wave(duration=1.5,start=1.5,framerate=48000)
c=t.SinSignal(freq=310,amp=1,offset=0)
c_wave=c.make_wave(duration=1.5,start=1.5,framerate=48000)
d=t.SinSignal(freq=200,amp=1,offset=0)
d_wave=d.make_wave(duration=1.5,start=1.5,framerate=48000)
e=t.SinSignal(freq=100,amp=1,offset=0)
e_wave=e.make_wave(duration=1.5,start=1.5,framerate=48000)
f=t.SinSignal(freq=170,amp=1,offset=0)
f_wave=f.make_wave(duration=1.5,start=1.5,framerate=48000)

music_tone=ab_wave+b_wave+c_wave+d_wave+e_wave+f_wave
from IPython.display import Audio
audio=Audio(data=music_tone.ys,rate=music_tone.framerate)
audio
```
